```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://ptv.com.pk/ptvWorld")
```

```python
soup = BeautifulSoup(response.content, 'html5lib')
headlines = soup.findAll("h5")
for data in headlines:
  headline = data.text.strip()
  print(headline)
```

```python
root = BeautifulSoup(response.content, 'html5lib')
print(root.prettify())
```
